```python
# Load the CSV file
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

#Convert the 'Date' column to datetime
df[' Date'] = pd.to_datetime(df[' Date'], dayfirst=True)

# Rename columns
df.rename(columns={
    'Region': 'State',
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate'
}, inplace=True)

# Preview the data
df.head()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(15,8))
corr = df[['Unemployment_Rate', 'Labour_Participation_Rate']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(15, 8))
sns.lineplot(data=df, x=' Date', y=' Estimated Unemployment Rate (%)', hue='State', legend=False)
plt.title("Unemployment Rate Over Time by State")
plt.xlabel(" Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.title("Indian Unemployment")
sns.histplot(data=df, x=' Estimated Unemployment Rate (%)', bins=20, kde=True)
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
lockdown_df = df[(df[' Date'] >= '2020-03-01') & (df[' Date'] <= '2020-06-30')]

plt.figure(figsize=(15, 8))
sns.lineplot(data=lockdown_df, x=' Date', y=' Estimated Unemployment Rate (%)', hue='State')
plt.title("Unemployment Rate During COVID-19 Lockdown (MarJun 2020)")
plt.xlabel(" Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
top_states = df.groupby('State')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).head(10)
top_states.plot(kind='barh', figsize=(10,6), color='tomato')
plt.xlabel("Average Unemployment Rate (%)")
plt.title("Top 10 States with Highest Average Unemployment Rate")
plt.gca().invert_yaxis()
plt.show()
```
